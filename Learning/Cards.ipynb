```python
import random

suits = ["Spades", "Hearts", "Diamonds", "Clubs"]
numbered = [2, 3, 4, 5, 6, 7, 8, 9, 10]
faces = ["Jack", "Queen", "King", "Ace"]
```

```python
def draw():
  suit = random.choice(suits)
  card = random.choice(numbered+faces)
  return card, "of", suit
```

```python
print(draw())
```

```python
print(draw())
```

```python
for _ in range(5):
  print(draw())
```
